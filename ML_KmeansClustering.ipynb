# K-Means Clustering


#### import libraries

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

#### Getting Data

```python
from sklearn.datasets import make_blobs
```

```python
df= make_blobs(n_samples= 200, n_features=2, centers=4, cluster_std=1.8, random_state=101)
```

#### Visualisation of Data

```python
plt.scatter(df[0][:,0],df[0][:,1], c=df[1],cmap='rainbow')
```

#### Clustering

```python
from sklearn.cluster import KMeans
```

```python
km= KMeans(n_clusters=4)
```

```python
km.fit(df[0])
```

```python
km.labels_
```

#### Comparing through Visualisation

```python
f, (ax2,ax1)= plt.subplots(1,2, sharey=True)
ax1.scatter(df[0][:,0],df[0][:,1],c=km.labels_, cmap='rainbow')
ax1.set_title("K means Clustering")
ax2.scatter(df[0][:,0],df[0][:,1],c=df[1])
ax2.set_title("Original Data")
```
